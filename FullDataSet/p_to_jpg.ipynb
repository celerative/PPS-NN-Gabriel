```python
Keras no soporta imagenes .p por lo que se las convertira en un paso previo a .jpg
```

```python
import matplotlib.pyplot as plt
import scipy
import csv
from scipy.misc import imsave


rootpath = './GTSRB/Training/Images'
outpath = './Data/Train/'


# loop over all 42 classes
for c in range(0,43):
    subdir = rootpath + '/' + format(c, '05d') + '/' # subdirectory for class
    gtFile = open(subdir + 'GT-'+ format(c, '05d') + '.csv', 'r') # annotations file
    gtReader = csv.reader(gtFile, delimiter=';') # csv parser for annotations file
    next(gtReader) # skip header
    
    wfile = open(outpath + format(c, 'd') + '/info.csv', 'w', newline='')
    writer = csv.writer(wfile, delimiter=';')
    writer.writerow(('FileName', 'Label'))
    
    # loop over all images in current annotations file
    i = 0
    for row in gtReader:
        image = plt.imread(subdir + row[0]) # lee p imagen
        imsave(outpath + format(c, 'd') + '/' + format(i, 'd') + '.jpg', image) #guarda jpg imagen
        writer.writerow((format(i, 'd') + '.jpg', row[7])) # primer nivel nombre de image, segundo label
        i+=1
    gtFile.close()
    wfile.close()
```

```python
rootpath = './GTSRB/Test/Images'
outpath = './Data/Test/'


# loop over all 42 classes
for c in range(0,43):
    subdir = rootpath + '/' # subdirectory for class
    gtFile = open(subdir + 'GT-final_test.csv', 'r') # annotations file
    gtReader = csv.reader(gtFile, delimiter=';') # csv parser for annotations file
    next(gtReader) # skip header
    
    wfile = open(outpath + format(c, 'd') + '/info.csv', 'w', newline='')
    writer = csv.writer(wfile, delimiter=';')
    writer.writerow(('FileName', 'Label'))
    
    # loop over all images in current annotations file
    i = 0
    for row in gtReader:
        if (int(row[7]) == c):
            image = plt.imread(subdir + row[0]) # lee p imagen
            imsave(outpath + format(c, 'd') + '/' + format(i, 'd') + '.jpg', image) #guarda jpg imagen
            writer.writerow((format(i, 'd') + '.jpg', row[7])) # primer nivel nombre de image, segundo label
            i+=1
    gtFile.close()
    wfile.close()
```
